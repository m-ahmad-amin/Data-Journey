```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

#### DataFrame

```python
df = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6],
    "score": [50, 55, 65, 70, 80, 90],
    "gender": ["M", "F", "M", "F", "M", "F"]
})

df
```

### Without hue

```python
sns.scatterplot(data=df, x="hours", y="score")
plt.show()
```

### With hue

```python
sns.scatterplot(data=df, x="hours", y="score", hue="gender")
plt.show()
```
