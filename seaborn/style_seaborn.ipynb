```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6],
    "score": [50, 55, 65, 70, 80, 90],
    "gender": ["M", "F", "M", "F", "M", "F"]
})
```

### Without Style

```python
sns.scatterplot(data=df, x="hours", y="score", hue="gender")
plt.show()
```

### With Style

```python
sns.scatterplot(
    data=df,
    x="hours",
    y="score",
    hue="gender",
    style="gender"
)

plt.show()
```
