```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
 
# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
 
plt.scatter(study_hours, exam_scores)
plt.annotate(
    "Peak Value", 
    xy=(3, 50),          # Arrow tip point
    xytext=(2.2, 28),    # Text location
    arrowprops=dict(
        arrowstyle="->",
        # facecolor='red', 
        # shrink=0.05,     # Moves arrow slightly away from text/point
        # width=20,         # Width of the arrow stem
        # headwidth=8      # Width of the arrowhead base
    )
)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
```
